```python
import numpy as np # except for numpy
```

define sigmoid functions

```python
# sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# gradient function
def sigmoid_deriv(x):
    return x * (1 - x)
```

create a dictionary to convert decimal to binary

```python
int2binary = {}
binary_dim = 8
 
largest_number = pow(2,binary_dim)
binary = np.unpackbits(
np.array([range(largest_number)],dtype=np.uint8).T,axis=1)
for i in range(largest_number):
    int2binary[i] = binary[i]
int2binary
```

define hyperparameters

```python
learning_rate = 0.1
n_x = 2
n_a = 16
n_y = 1
```

initialize weights

```python
np.random.seed(1)
w0 = 2*np.random.random((n_x, n_a)) - 1
w1 = 2*np.random.random((n_a, n_y)) - 1
wa = 2*np.random.random((n_a, n_a)) - 1
 
dw0 = np.zeros_like(w0)
dw1 = np.zeros_like(w1)
dwa = np.zeros_like(wa)
```

```python
np.random.seed(1)
for i in range(10000):
    a_int = np.random.randint(largest_number/2) # int version
    a = int2binary[a_int] # binary encoding
     
    b_int = np.random.randint(largest_number/2) # int version
    b = int2binary[b_int] # binary encoding
    
    c_int = a_int + b_int
    c = int2binary[c_int]
    
    d = np.zeros_like(c)
    
    overallError = 0
    
    dy_ = list()
    Z1 = list()
    Z1.append(np.zeros(n_a))
    
    for j in reversed(range(binary_dim)):
        X = np.array([[a[j],b[j]]])
        y = np.array([[c[j]]]).T
    
        z1 = sigmoid(X.dot(w0) + Z1[-1].dot(wa))
        
        y_ = sigmoid(z1.dot(w1))
        
        error = y - y_
        dy_.append(error*sigmoid_deriv(y_))
        overallError += abs(error[0])
        
        d[j] = np.round(y_[0][0])
        
        Z1.append(np.copy(z1))
    
    dz1_next = np.zeros(n_a)
    
    for j in range(binary_dim):
        X = np.array([[a[j],b[j]]])
        z1 = Z1[-j-1]
        z1_prev = Z1[-j-2]
        
        dy = dy_[-j-1]
        dz1 = (dz1_next.dot(wa.T) + dy.dot(w1.T)) * sigmoid_deriv(z1)
    
        dw1 += np.atleast_2d(z1).T.dot(dy)
        dwa += np.atleast_2d(z1_prev).T.dot(dz1)
        dw0 += X.T.dot(dz1)

        dz1_next = dz1
    
    w0 += dw0 * learning_rate
    w1 += dw1 * learning_rate
    wa += dwa * learning_rate
    
    dw0 *= 0
    dw1 *= 0
    dwa *= 0
    
    if(i % 1000 == 0):
        print("Error:" + str(overallError))
        print("Pred:" + str(d))
        print("True:" + str(c))
        out = 0
        for index,x in enumerate(reversed(d)):
            out += x*pow(2,index)
        print(str(a_int) + " + " + str(b_int) + " = " + str(out))
        print("------------")
```
